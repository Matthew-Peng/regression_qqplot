```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

### Generate data

```python
n = 1000
e = np.random.normal(0,1,n)
# X = np.random.beta(1,1,n)
X = np.random.random_sample(n)
y = 10 + 5*X + e
```

```python
plt.hist(y)
```

### Fit a linear regression model

```python
reg = LinearRegression()
reg.fit(X.reshape(-1, 1), y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

### Plot residual

```python
y_hat = reg.predict(X.reshape(-1, 1))
residual = y - y_hat
plt.plot(residual)
```

```python
fig = plt.hist(residual, bins=100)
plt.show()
```

### Plot qqplot on residual

```python
import statsmodels.api as sm
fig = sm.qqplot(residual)
plt.show()
```

### Plot qqplot on y

```python
fig = sm.qqplot(y)
plt.show()
```

# What if e is not normal distribution

```python
n = 1000
e = np.random.random_sample(n)
# X = np.random.beta(1,1,n)
X = np.random.random_sample(n)
y = 10 + 5*X + e
```

```python
plt.hist(y)
```

```python
reg = LinearRegression()
reg.fit(X.reshape(-1, 1), y)
print(reg.coef_)
print(reg.intercept_)
```

### Plot residual

```python
fig = plt.hist(residual, bins=100)
plt.show()
```
